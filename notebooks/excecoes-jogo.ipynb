## Testes da Hierarquia de Exceções

Monte um código que teste a sua hierarquia exceções (todas elas) conforme o exemplo a seguir.

```java
public class ErroArquivo extends Exception {
   public ErroArquivo() {
      super();
   }

   public ErroArquivo(String m) {
      super(m);
   }
}
```

```java
public class ClassNotFound extends ErroArquivo {
   public ClassNotFound() {
      super();
   }

   public ClassNotFound(String s) {
      super(s);
   }
}
```

```java
public class NullPointer extends ErroArquivo {
   public NullPointer() {
      super();
   }

   public NullPointer(String e) {
      super(e);
   }
}
```

```java
import java.io.ObjectInputStream;
import java.io.File;
import java.io.FileInputStream;
public class Arquivo {
   
    public static ObjectInputStream entrada(ObjectInputStream o) throws ErroArquivo {
        ObjectInputStream ent;
        if (o == null)
            throw new NullPointer ("Ponteiro nulo");
        if (o.getClass() != ObjectInputStream.class)
            throw new ClassNotFound ("Classe nao encontrada");
        File arquivo = new File("leaderboard.ser");
        ent = new ObjectInputStream (new FileInputStream(arquivo));
        return ent;
    }
}
```
